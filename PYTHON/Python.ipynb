# 1. Prepare your data for analysis and then proceed And list down your findings

```python
import pandas as pd
```

```python
df= pd.read_csv('raw_data.csv')
df
```

```python
df = df.loc[:, ~df.columns.str.contains('^Unnamed')]
df
```

```python
df.rename(columns=  {"location":"COUNTRY"}, inplace=True)
df.rename(columns=  {"iso_code":"CODE"}, inplace=True)
df.rename(columns=  {"date":"DATE"}, inplace=True)
df.rename(columns=  {"total_cases":"TC"}, inplace=True)
df.rename(columns=  {"total_deaths":"TD"}, inplace=True)
df.rename(columns=  {"stringency_index":"STI"}, inplace=True)
df.rename(columns=  {"population":"POP"}, inplace=True)
df.rename(columns=  {"gdp_per_capita":"GDPCAP"}, inplace=True)
df.rename(columns=  {"human_development_index":"HDI"}, inplace=True)
```

```python
df=df[["CODE", "COUNTRY", "DATE", "HDI", "TC","TD","STI","POP","GDPCAP"]]
df
```
